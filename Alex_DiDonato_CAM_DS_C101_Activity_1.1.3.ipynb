# Google Flu Trends (GFT)

Google launched a data project to predict flu outbreaks by analysing search query patterns. The hypotheses was that an increase in searches for flu symptoms or treatments for flu would correlate with an increase of the virus in that area.

This is cited as a textbook case of big data analysis failure.

The analysis incorrectly focussed solely on search behaviour of users, which turned out to be overly simplistic and too narrow a focus. 

Models were trained on data from search terms from a specific period of time (2003 - 2008), where good initial correlation was observed between search query frequency and flu incindence. However the models created drifted and because they focussed solely on search popularity which often reflected media coverage or public anxiety, not actual illness.

## The cause:

The wrong kind of data was used for the analytics being sought.

Several key questions were missed when considering the data used in the analysis and model:

- Instead of how can we use searches related to flu to predict flu outbreaks? There should have been a consideration for what are the causes of people searching about flu? This was not fully explored.
- What other data sources can be used to improve model robustness?
- How can we regularly check validate and refine the models that we use?

## How could it be improved?

If I had been involved I would have broadened the analytical focus and looked to combine the search history with other data sources:

- Verified data on actual flu occurences
- Social media

I would have asked the questions: 

- What causes people to search about the flu? 
- How can search data complement medical evidence to track flu accurately?

Instead of trying to model solely against prevalence of search terms.
