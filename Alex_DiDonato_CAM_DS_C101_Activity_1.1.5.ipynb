# Activity 1.1.5: Using statistical thinking to analyse and define a business problem from a given scenario

## Scenario
**Communik8** is a telecommunications company with both online and physical stores. It provides mobile contracts and sells handsets. Recently, the company has experienced a sizeable reduction in new customers and a decrease in the retention of existing customers.

With the primary goal of improving business performance, the company needs to identify critical reasons for this recent decline. The company would like to make sense of its customer data and transform it into actionable insights.

The management team at Communik8 has provided recent customer feedback to your team of data scientists and would like initial feedback on the main issues that can be identified from this data.

## Objective
The goal is to explore and analyse an organisational scenario to identify key problems.

## Assessment criteria
By completing this activity, you will be able to provide evidence that you can:
* Note common themes and sentiments expressed by customers without performing any in-depth analysis.
* Formulate your own business problem statement.


## Activity guide

3. Identify how solving these problems will help the company achieve its objectives and goals and outline the consequences of not achieving these goals (e.g. improved customer satisfaction, customer retention, loss of revenue, and so on).

## Problem statement structure:

- Problem Identified
- Impact of problem on busniness
- Objectives to solving the problem

The company is facing a decline in customer satisfaction and operational efficiency. Long call wait times, delayed deliveries, poor network reliability, inconsistent in-store service, and unresponsive online support are reducing customer satisfaction and trust. Additionally, billing inaccuracies, failed cancellations, and insufficient staff training have compounded the problem, leading to increased complaints and operational strain. This is leading to, cancelled contracts, and lost revenue, increased support costs, and unecessary operation overhead (customer support, rectifying mistakes and incorrect revenue protection actions) and damage to the brand’s reputation which will ultimately lead to fewer sales and reduced revenue. Existing customers are losing confidence in both digital and in-store channels, which will lead to higher customer churn.

The company needs to identify the root causes of customer dissatisfaction resulting from, delivery inefficiencies, process failures, network reliability and decline in customer service in the form of, a poor online / digital experience, long call wait times and unhelpful customer service staff. The company should implement targeted strategies for improvement, to strengthen these areas.

Getting business context correct is key to identifying issues and problems, and tying problems to business objectives allows business consequences to be predicted and therefore a strong problem statement to be constructed.
