# 1. Problem Statement (from 1.1.5)

The company is facing a decline in customer satisfaction and operational efficiency. Long call wait times, delayed deliveries, poor network reliability, inconsistent in-store service, and unresponsive online support are reducing customer satisfaction and trust. Additionally, billing inaccuracies, failed cancellations, and insufficient staff training have compounded the problem, leading to increased complaints and operational strain. This is leading to, cancelled contracts, and lost revenue, increased support costs, and unecessary operation overhead (customer support, rectifying mistakes and incorrect revenue protection actions) and damage to the brand’s reputation which will ultimately lead to fewer sales and reduced revenue. Existing customers are losing confidence in both digital and in-store channels, which will lead to higher customer churn.

The company needs to identify the root causes of customer dissatisfaction resulting from, delivery inefficiencies, process failures, network reliability and decline in customer service in the form of, a poor online / digital experience, long call wait times and unhelpful customer service staff. The company should implement targeted strategies for improvement, to strengthen these areas.

## 2. Hypothesis Statement

Variables covered:
- Online Support Use Success 
    - Measured with:
        - Completion rate of self-service interactions
        - Chat satisfaction scores
        - Abandonment rates
- Call wait times

Hypothesis(1):

An increase in the number of Customers sucessfully using online resources, FAQs, and online chat to answer common customer queries, and / or log support requests, will result in fall in demand for customer support calls leading to shorter call queues and wait times. Conversly a poor customer experience when accessing online resources will result in a greater volume of customers resorting to contacting customer support directly by phone, where they will experience higher wait times due to the higher volume of calls. If we analyse customer use of the available online support resources and determine a success metric from this (e.g. an interaction would be deemed unsucessful if a customer engaged in online chat and subsequently abandoned the interaction), and compare this with call queue volumes and wait times, we can identify opportunities to reduce call volumes.

H₀: There is no relationship between successful online support interactions and call wait times.
H₁: Increased successful online support interactions are associated with reduced call wait times.


# 3. if-then

(1) If the number of customers successfully using online support resources (such as FAQs, self-service tools, and chat) to resolve their issues increases,
then the volume of customer support calls and the average call wait times will decrease; more effective online support reduces the need for customers to contact the call centre directly.

(2) If analysis confirms that customers who successfully use online support resources (such as FAQs, self-service tools, and chat) experience fewer needs to contact the call centre, resulting in reduced call volumes and shorter wait times, then enhancing the effectiveness of online support will directly reduce demand on customer support, improve customer satisfaction, and increase overall operational efficiency.
