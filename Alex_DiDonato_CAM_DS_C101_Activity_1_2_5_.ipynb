# 2. Features that will provide the best insights into customer behaviour.

- Repeat purchase rate

    A customer that has a high repeat purchase rate may indicate loyalty to the company. The metric could be defined as a Repeat Purchase Rate = (Number of customers with >1 purchase in period) รท (Total active customers in same period)
    
    The rate can be measured over time, and compared period to period to give an indication of overall customer loyalty over time.

    Segment by Customer age, Gender, Location: Grouping the metric by Customer age, Gender & Location will allow a per-cohort measurement and potentially inform where further marketing or loyalty schemes could be targeted.

- Products purchased

    Shows customer buying patterns - a diverse range of purchased products indicates a opportunities to sell bundles and cross-selling (e.g. adding 'customers also bought' recommendations to a website or discounts on bundles). Using a 'count of distinct product categories per customer' will allow a diversity index to be derived and used as a metric to measure cross-selling opportunity.

- Cart abandonment rate

    This feature reveals customer buying intent - the basket content indicates which products are of interest. Measuring abandonment by product type could signal pricing issues at a product / product type level, checkout experience, or issues with shipping / delivery.

    Abandonment rate can be measured per product on a weekly basis as 

$$
\text{Cart Abandonment Rate} = 1 - \left( \frac{\text{Number of Completed Purchases}}{\text{Number of Baskets Created}} \right)
$$

- Customer satisfaction score

    This feature is a measure of direct feedback from the customer - a high customer satisfaction score signals good customer engagement and should correlate with low churn / strong retention. Customer satisfaction per product type can identify areas where service or product experience varies.

- Loyalty membership

    Where a customer signs up for a loyalty membership this is could be an indication of intention to make repeat purchases.

    Use length of membership and engagement / use of membership benefits to create a metric that can be used to model customer behaviour.
